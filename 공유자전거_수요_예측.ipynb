### feature engineering
> 시간 데이터 변환

```python
bike_df['datetime']=pd.to_datetime(bike_df['datetime'])
```

```python
bike_df.info()
```

```python
bike_df.head(3)
```

```python
bike_df['datetime'].dt.year
```

```python
bike_df['datetime'].dt.month
```

```python
bike_df['datetime'].dt.hour
```

```python
bike_df['year']=bike_df['datetime'].dt.year
bike_df['month']=bike_df['datetime'].dt.month
```

```python
bike_df['hour'] = bike_df['datetime'].dt.hour
```

```python
bike_df.head(3)
```

```python
bike_df['date']=bike_df['datetime'].dt.date
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x='date',y='count',data=bike_df)
# plt.xticks(rotation = 45)
plt.show()
```

```python
bike_df[bike_df['year']==2019].groupby('month')['count'].mean()
```

```python
bike_df[bike_df['year']==2020].groupby('month')['count'].mean()
```

```python
bike_df[bike_df['year']==2021].groupby('month')['count'].mean()
```

```python
def covid(date):
    if str(date) < '2020-04-01':
        return 'precovid'
    elif str(date) < '2021-04-01':
        return 'covid'
    else:
        return 'postcovid'
```

```python
bike_df['date'].apply(covid)
```

```python
bike_df['covid'] = bike_df['date'].apply(lambda date: 'precovide' if str(date) < '2020-04-01' else 'covid' if str(date) < '2021-04-01' else 'postcovid')
```

```python
bike_df['season'] = bike_df['month'].apply(lambda x:'winter' if x ==12 else 'fall' if x >=9 else 'summer' if x >= 6 else 'spring' if x>=3 else 'winter')
```

```python
bike_df['day_night']=bike_df['hour'].apply(lambda x: 'night' if x>=21 else 'late evening' if x>=19 else 'early evening' if x >=17 else 'late afternoon' if x >= 16  else ' early afternoon' if x>=13 else 'late morning' if x>=11 else 'early morning' if x>=5 else 'night')
```

```python
bike_df.head(3)
```

```python
bike_df.drop(['datetime','month','date'],axis=1,inplace=True)
```

```python
bike_df.head(3)
```

### 더미변수 만들기

```python
for i in ['weather_main','hour','covid','season','day_night']:
    print(i,bike_df[i].nunique())
```

```python
bike_df['weather_main'].unique()
```

```python
bike_df=pd.get_dummies(bike_df, columns=['weather_main', 'hour', 'covid', 'season', 'day_night'])
```

### 모델링

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(bike_df.drop('count',axis=1),bike_df['count'],test_size=0.3, random_state = 100)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor(random_state=100)
```

```python
dt.fit(X_train,y_train)
```

```python
pred_1=dt.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=pred_1)
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(y_test,pred_1)
```

### 리니어 리그레션과 비교

```python
from sklearn.linear_model import LinearRegression
```

```python
pred_3
```

```python
pred_3 = np.exp(pred_3)-1
```

```python
root_mean_squared_error(y_test, pred_3)
```
