##  시행과 사건

확률을 이해하기 위해 먼저 **시행**과 **사건**에 대해 알아보자.

###  시행 (trial)
- **정의**: 반복할 수 있으며, 매번 결과가 달라질 수 있는 실험

    ex) 주사위 2개를 던지는 행위

###  사건 (event)
- **정의**: 시행에 따라 발생할 수 있는 **특정 결과나 조건을 만족하는 결과들의 집합**
    
    ex): 주사위 두 개의 눈의 합이 7이 되는 사건

## 확률 함수
>(Probability Function)

확률 변수는 시행할 때마다 달라질 수 있는 값(예: 주사위 눈의 합)을 나타낸다.  
**확률 함수**는 이러한 확률 변수의 각 값에 **확률을 부여하는 함수**이다.

확률 함수는 일반적으로 \( P \)로 표현되며,  
확률 변수 \( X \)가 특정 값 \( x \)를 가질 확률은 다음과 같이 쓴다:

$P(X = x)$


ex)
주사위 두 개를 던지는 시행에서, 눈금의 합이 3이 될 확률은:

$P(X=3)= 1/36$

### 확률 변수
> (Random Variable)

확률 변수란, 사건에 따라 그 값이 **확률적으로 정해지는 변수**를 의미한다.


주사위 2개를 던지는 시행에서, 눈금의 합은 시행마다 달라질 수 있다.  
따라서 이 상황에서는 다음과 같이 표현할 수 있다:

**확률 변수 ( $X$ ) = 주사위 두 눈의 합**

예를 들어, 한 번 던졌을 때 나올 수 있는 눈금의 합은 2부터 12까지이며, 이는 확률적으로 결정된다.

### 표기법
- 확률 변수: 대문자 ( $X$ ) 로 표기  
- 확률 변수가 가질 수 있는 값: 소문자 \( $x$ \) 로 표기  

<br>

<br>

### 확률 변수 예시

앞면에 1, 뒷면에 0이 쓰여있는 2개의 동전이 있다고 하자.

- 시행: 동전 2개를 동시에 던지기  
- 확률 변수 \( $X$ \): 두 눈금의 합

이때 가능한 \( $X$  \)의 값은 0, 1, 2이다.

시행할 때마다 \( $X$  \) 값이 달라질 수 있으므로, 확률적으로 변할 수 있다 하여  
\( $X$  \)를 **확률 변수**라 한다.

---

ex)  
- 동전 1: 앞면(1), 동전 2: 앞면(1) → \( $X = 1 + 1 = 2$ \)  
- 동전 1: 앞면(1), 동전 2: 뒷면(0) → \( $X = 1 + 0 = 1$ \)  
- 동전 1: 뒷면(0), 동전 2: 뒷면(0) → \( $X = 0 + 0 = 0$ \)

### 확률 변수는 함수다?

- **표본공간 (Sample Space)**:  
  발생 가능한 모든 사건들의 집합. 즉, 시행의 모든 가능한 결과를 포함하는 **전체 집합**을 의미한다.  

- **확률 변수 (Random Variable)**:  
  표본공간 안에서 발생하는 사건을 특정 **수치로 대응시키는 함수**이다.  
  즉, 확률 변수는 다음과 같은 함수로 이해할 수 있다:

  $$
  X: \Omega \rightarrow \mathbb{R}
  $$

  여기서 \( X($\Omega$) \)는 표본공간의 원소 \( $\Omega$ \)에 대응되는 수치 값을 의미한다.

- **상태 공간 (State Space)**:  
  확률 변수 \( 𝓧 \)가 취할 수 있는 모든 값 \( 𝑥 \)들의 집합을 말한다.  
  이는 \( 𝓧 \)의 **출력값의 집합**, 즉 정의역이 아닌 **공역**이다.

---

 ex)
- 주사위 한 개를 던지는 경우:  
  - 표본공간 \( $\Omega$ = \{1, 2, 3, 4, 5, 6\} \)  
  - 확률 변수 \( 𝓧 \): 나오는 눈의 수  
  - 상태 공간 \( \$\mathcal{X} $ = {1, 2, 3, 4, 5, 6\} \)

* 확률 분포를 통해 통계적인 특성을 쉽게 이해할 수 있다.

### 확률 분포함수

확률 분포함수는 **확률변수 $X$가 가지는 값 $x$에 확률 $P(X = x)$를 대응시키는 함수**를 의미합니다.  
즉, 어떤 값 $x$가 나타날 확률을 알려주는 함수입니다.

> [참고]  
> 확률 분포 그 자체를 함수로 보기도 하며, 이를 **확률 분포 함수 $P$**라고 부르기도 합니다.

확률 분포 함수는 크게 다음 두 가지로 나뉩니다:

1. **확률질량함수 (Probability Mass Function, PMF)**  
   - 이산 확률변수(discrete random variable)에 대한 확률 분포 함수입니다.
   - 예: 주사위 눈금, 동전 던지기 결과 등

2. **확률밀도함수 (Probability Density Function, PDF)**  
   - 연속 확률변수(continuous random variable)에 대한 확률 분포 함수입니다.
   - 예: 키, 몸무게, 온도 등

또한, **모든 사건에 대하여 확률 분포 함수의 값을 나열한 것**을 **확률 분포 (Probability Distribution)**라고 합니다.

# 이산확률분포

## 이산확률분포

>(Discrete Probability Distribution)

* 확률변수 $X$가 취할 수 있는 모든 값을 셀 수 있는 경우, 이를 **이산확률변수**라고 한다.
* 이때 **이산확률분포**는 이산확률변수의 확률 분포를 의미한다.
* 주사위를 던졌을 때 나올 수 있는 눈금은 {1,2,3,4,5,6} 중 하나이다.
* 따라서, 이 경우 "주사위 눈금"의 값은 6개만 존재하므로, 이산확률변수이다.

<br>

---

<br>

###  이산 확률분포 예시

- 주사위를 던졌을 때 나올 수 있는 눈금(수)을 확률변수 $X$ 라고 하자.
- 확률변수 $X$는 다음의 이산값들 중 하나를 가질 수 있다:  
  $$
  X \in \{1, 2, 3, 4, 5, 6\}
  $$
- 각 눈금이 나올 확률은 모두 동일하며, 다음과 같다:  
  $$
  P(X = 1) = P(X = 2) = \cdots = P(X = 6) = \frac{1}{6}
  $$

> 이때 주사위 눈금 $X$에 대한 확률 분포는 **이산 균등 분포 (Discrete Uniform Distribution)**이다.

###  이산 균등 분포란?

모든 가능한 값이 **동일한 확률**로 나타나는 이산 확률 분포를 의미합니다.

- 일반적인 형태:
  $$
  P(X = x_i) = \frac{1}{n}, \quad x_i \in \{x_1, x_2, ..., x_n\}
  $$
- 주사위 예시에서 $n = 6$ 이므로, 각 눈금이 나올 확률은 $\frac{1}{6}$입니다.

 ## 확률질량함수
 >( Probability Mass Function , PMF)

*  **확률질량함수**는 이산확률 변수가 특정한 값을 가질 **확률을 출력하는 함수**다.
* **확률질량함수**는 이산확률분포를 표현하기 위해 사용하는 확률분포함수로 이해할 수 있다.
* 동전 2개를 동시에 던지는 시행에서 두 눈금의 합을 $X$라고 하자.
* 이때, $X$는 이산확률변수로, 확률질량함수 $f(x)$는 다음과 같이 정의할 수 있다.

$$
\ f(0) = P( X = 0 ) = 1/4 \
$$
$$
\ f(1) = P( X = 1 ) = 1/2
$$
$$
f(2) = P( X = 2 ) = 1/4
$$

* 확률 변수 X에 대한 확률 질량 함수라는 의미로 $fₓ(x)$라고 표기하기도 한다.

## 베르누이 확률변수와 베르누이 확률분포

---

**1. 베르누이 확률변수 (Bernoulli Random Variable)**

- 결과가 **두 가지**(성공/실패)뿐인 실험에서 정의되는 확률변수
- 확률 변수는 0 혹은 1의 값만 가질 수 있으므로, 이산확률변수다.
- 일반적으로 값은:
  - 성공: 1
  - 실패: 0
- 확률변수 $X$는 다음을 만족함:

$$
X =
\begin{cases}
1 & \text{with probability } \mu \\
0 & \text{with probability } 1 - \mu
\end{cases}
$$

---

**2. 베르누이 확률분포 (Bernoulli Distribution)**

- 베르누이 확률변수가 따르는 분포
- 매개변수 $\mu$는 **성공할 확률(1이 나올 확률)**, $0 \le \mu \le 1$
- 확률질량함수(PMF):

$$
P(X = x) = \mu^x (1 - \mu)^{1 - x}, \quad x \in \{0, 1\}
$$

---

3. 표기법: $X \sim \text{Bern}(x; \mu)$

- 이 표기법은 아래 두 가지 의미로 나뉨:

| 표기                                   | 의미 |
|-------------------------------------|------|
| $$
    X \sim \text{Bernoulli}(\mu)
   $$     | 확률변수 $X$는 성공 확률 $\mu$를 가지는 베르누이 분포를 따른다 |
| $$
\text{Bern}(x; \mu)
$$               | $X = x$일 때의 확률 값을 의미한다, 즉 $P(X = x)$ |

###  예시:

- $X \sim \text{Bernoulli}(0.7)$  
  → $X$는 성공 확률이 0.7인 베르누이 확률변수

- $\text{Bern}(1; 0.7) = 0.7$  
  → $P(X = 1) = 0.7$

- $\text{Bern}(0; 0.7) = 0.3$  
  → $P(X = 0) = 0.3$

**[참고]**

* 모수(parameter)는 세미콜론(;) 기호로 구분하여 표기한다.
* 베르누이 확률분포는 모수로 $\mu$를 가지는데, 1이 나올 확률을 의미한다.

<br>

## 베르누이 분포의 확률질량함수
> (Bernoulli Distribution PMF)

- 베르누이 확률 분포의 확률질량함수는 다음과 같다:
$$
Bern(x; \mu) =
\begin{cases}
\mu & \text{if } x = 1 \\
1 - \mu & \text{if } x = 0
\end{cases}
$$
- 이는 간단히 아래와 같은 하나의 수식으로도 표현할 수 있다:

$$
Bern(x; \mu) = \mu^x (1 - \mu)^{1 - x}
$$

여기서:

- $( x \in \{0, 1\} )$
- \( $\mu$ \)는 사건이 발생할 확률 (즉, 성공 확률)



---


> $\mu$ 가 0.8인 베르누이 확률 분포는 다음과 같다.

$$
Bern(x; \mu) =
\begin{cases}
\mu & \text{if } x = 1 \\
1 - \mu & \text{if } x = 0
\end{cases}
$$

## 이항분포 개요

> 베르누이 시행을 N번 반복하는 경우가 있다.

ex) 동전던지기 7번 시행

<br/>

---  
<br/>

###이항분포란?
* 성공확률이 $\mu$ 인 베르누이 실행을 $N$번 반복한다.
* $N$번 중에서 성공한 횟수를 확률변수 $X$라고 하자.
* $X$는 0부터 $N$까지의 정수 중 하나이다.
* 이러한 확률 변수를 이항 분포를 따른다고 한다.
    <br>
        $X \sim \text{Bin}(x;N, \mu)$
    <br><br>

* 이항 분포는 모수( parameter)로 $N$과 $\mu$를 가진다.
* 파라미터 1: 시행 횟수 $N$
* 파라미터 2: 한 번의 횟수에서 1이 나올 확률 $\mu$


<br/>

---  
<br/>

### 이항 분포의 확률질량함수 (Binomial Distribution PMF)

- 이항 분포 확률 변수 \( X \)의 확률질량함수는 다음과 같다:

$$
X \sim Bin(x; N, \mu) = \binom{N}{x} \mu^x (1 - \mu)^{N - x}
$$

- 단, \( $\binom{N}{x}$ \)는 \( $N$ \)개에서 \( $x$ \)개를 선택하는 조합(combination)의 수와 같다.

- 단, \$( N! = N \cdot (N - 1) \cdot \dots \cdot 2 \cdot 1 $\)이다.

## 포아송 분포
>(Poisson Distribution)

포아송 분포는 **단위 시간 또는 단위 공간에서 사건이 일정한 평균 빈도로 발생**할 때,  
해당 구간 안에서 사건이 **몇 번 발생하는지를 나타내는 이산 확률 분포**이다.

---

###  확률질량함수 (PMF)

$$
P(X = x) = \frac{\lambda^x e^{-\lambda}}{x!}, \quad x = 0, 1, 2, \dots
$$

- $( X )$: 단위 시간/공간 내에 발생한 사건 수  
- $( λ)$: 평균 발생 횟수  
- $( e )$: 자연상수 (\( ≈ 2.718 \))  
- $( x!)$: x의 계승

---

###  성질

- **기대값 (평균)**: $( \mathbb{E}[X] = λ $)
- **분산**: $( \text{Var}(X) = \lambda $)  
- **비중첩 구간은 서로 독립**

---

###  예시

> 한 시간에 평균 4명이 고객센터에 전화를 건다고 할 때,  
> 한 시간 동안 정확히 2명이 전화할 확률은?

$$
P(X = 2) = \frac{4^2 e^{-4}}{2!} = \frac{16 e^{-4}}{2}
$$

---

###  사용 예시

- 시간당 발생하는 고객 콜 수  
- 하루 동안 발생하는 사고 건수  
- 일정 구간 내 차량 통과 수  
- 유전자 돌연변이 발생 수

* 단위 시간 내 **평균 발생 횟수($\lambda$)가 5일때, 그 사건이 8회 일어날 확률은?

$$
 \frac{5^8 e^{-5}}{8!} = 0.06528
$$

# 연속확률분포

## 확률밀도함수
> (Probability Density Function)

* 연속확률변수가 주어진 구간 내에 포함될 확룰을 출력하는 함수다.

**[참고]** 이산확률변수는 확률밀도함수(PDF)가 아닌 확률질량함수(PMF)를 사용한다.

<br>

---
<br>

## 연속확률분포의 성질

1. 확률 변수 $X$가 어떠한 구간에 속할 확률은 0과 1사이이다.
2. 확률 변수 $X$가 값을 가질 수 있는 모든 구간의 확률을 합치면 1이다. (전체면적 = 1)
→ 단, 각 구간은 배반(서로 겹치는게 없을 때) 관계일 때 이것이 성립한다.

image.png

* 확률밀도함수를 어떻게 이해할 수 있을까?
* $𝒇(x)$는 확률을 의미하고 $dx$는 구간 길이를 의미한다.
* 다시 말해 $α$부터 $b$까지의 구간에 대하여 "확률 x 구간길이"의 값을 모두 더한 값이다.

$$
P(α≤ X ≤ b) = ∫_a^bf(x)\,dx
$$

<br>

<br>

## 연속확률분포- 참고사항

* 확률변수 $X$의 값이 구간 $[a,b]$에 속할 확률은 다음과 같이 표현한다.

$$
P(α≤X ≤b)
$$

* 연속 확률 변수는 면적으로 계산되며, 한 점에 대한 확률은 0으로 간주한다.
* 예시) 나랑 키가 소수점 아래까지 "완벽히" 동일한 사람이 존재할 수 있을까?

$$P(α≤X ≤b) = P(α < X < b)$$

##균등분포
>(Uniform Distribution)

* 가장 단순한 연속확률분포로, 특정 구간 내 값들이 나타날 가능성이 균등하다.
* 다시 말해, 모든 확률변수에 대해 일정한 확률을 가지는 확률 분포다.
* X가 균등 분포를 따를 때 $X ~ U(a,b)$로 표현한다.
* X는 α에서 b 사이에서 일정한 값을 취하고, $P(α < X < b) = 1$이다.
* 균등 분포를 따르는 확률 변수 X의 확률밀도함수는 다음과 같다.
<br><br>
$$
f(x) =
\begin{cases}
\frac{1}{b-a} & \text{α≤ X ≤ b } \\
0 & \text{otherwise }
\end{cases}
$$

## 균등분포 예시 - 스팸메일

* 매일 오후 3시 ~ 오후 4시 사이에 하나의 스팸 메일이 도착.
* 해당 시간 안에서 특정 시간에 스팸메일이 도착할 확률이 동일하다고 할때,
* 다시 말해 오후 3시~ 오후 4시 사이에서 균등한 확률 값을 가진다.
* 이때 스팸 메일이 도착하는 시각을 확률 변수 X라고 해보자.
* 확률 분포 함수는 어떻게 그릴 수 있을까?

> 균등분포는 **직사각형 형태**를 보인다.<br>
가로축의 단위를 "분"으로 했을때, 가로 길이가 60이 된다.<br>
이때 오후 3시부터 오후 4시 시간대의 각 분에 대하여 $𝑭(x)$는 1/60이다.

image.png

**[문제]** 확률변수 X가 구간 [10,20]에서 균등한 분포를 가질 때, $P(15≤X ≤17)$은?

**[풀이]** X의 확률 밀도 함수가 $f(x) = \frac{1}{20-10} = \frac{1}{10}$이므로, $P(15≤X≤17) = \frac{1}{10}$ 𝖷 $2 = \frac{1}{5}$이다.

## 이산 균등 분포

> Discrete Uniform Distribution

* **[참고]** 균등 분포는 이산확률변수에 대해서도 정의될 수 있다.
* 이산 균등분포(discrete uniform distribution)의 예시를 확인해 보자.
* 예를 들어 주사위를 한번 던지는 시행에서, 주사위의 눈금 값을 확률 변수 $X$라고 해보자.

image.png

> 즉, 확률변수 X에 대해서 모두 위와 같이 같은 확률 값을 가지며, 이산확률분포일 때에도 이산적인 균등분포가 나올 수 있기 때문에 이산균등분포라고 표현할 수 있다.

## 정규분포

> Normal Distribution

* 정규 분포는 기계학습 분야에서 매우 자주 등장함
* 정규 분포는 밥그릇과 같은 모양을 보임

image.png

### 정규분포의 특징

* 정규 분포의 모양은 ①평균과 ②표준편차로 결정된다.
* 확률밀도함수는 평균을 중심으로 좌우 대칭인 종 모양을 형성한다.
* 관측되는 값의 약 98%가 $±2σ$ 범위 안에 속한다.

image.png

### 정규분포의 특징 - 평균

* 평균에 따라서 정규 분포가 좌우로 **평행이동**한다.
* 평균이 0인 정규 분포와, 평균이 2인 정규 분포를 비교할 수 있다.

image.png

### 정규 분포의 특징 - 분산

* 분산이 클수록 정규 분포가 옆으로 넒게 퍼지게 된다.
* 분산이 작을수록 정규분포는 가파른 모양을 가지게 된다.

## 정규 분포의 확률밀도함수

* 공학 분야에서는 가우시안(Gaussian)분포로 부르기도 한다.
* 확률 변수 X의 확률밀도함수가 다음과 같을 때, X가 정규분포를 따른다고 한다.
* 평균 $μ$과 분산 $𝛔^2$에 의해 분포의 모양이 결정된다. $(-∞< x <∞; σ>0)$

<br><br>
$$
f(x) = \frac{1}{\sqrt{2\pi}\,\sigma} \cdot e^{-\frac{1}{2\sigma^2}(x - \mu)^2}
$$


> 현실 세계의 많은 데이터의 상당수가 정규 분포로 매우 가깝게 표현되는 경향이 있다.
<br> 그래서 현실 세계에서 수집된 데이터의 분포를 근사할 때 자주 사용된다.

## 표준 정규 분포

> Normal Distribution

* 표준 정규 분포는 평균이 0, 표준편차가 1인 정규분포를 의미한다.
* 확률을 계산하기 위해 정규 분포 함수를 직접 적분하는 것은 매우 어렵고 번거롭다.
* 길제로는 정규 분포 한ㅁ수를 표준 정규 분포로 변환한 뒤에 확률을 계산한다.

## 지수분포
> Exponential Distribution

* 특정 시점에서 어떤 사건이 일어날 때까지 걸리는 시간을 측정할 때 사용한다.
* 웹 사이트에 평균적으로 10분에 한 명씩 방문자가 접속한다.
* **[문제]** 한 명의 방문자가 접속한 뒤에, 그 다음 방문자가 올 때까지 걸리는 시간의 확률을 구해보자.

* 앞서 포아송 분포는 **발생 횟수**에 대한 확률을 구할 때 사용했다.
* 지수 분포는 **대기시간**에 대한 확률을 구할 때 사용한다.

image.png

* 지수 분포의 확률 밀도 함수는 다음과 같다.
* $λ$: 단위 시간 동안 평균 사건 발생 횟수

$$
    f(x) = λe^{-λx},x≥0
$$

## 지수 분포 예시 - 해킹 시도

**[문제]**

* 운영 중인 서버에는 하루 평균 4건의 해킹이 시도된다.
* 첫 번째 해킹 시도가 3시간 안에 발생할 확률은?

**[해설]**

* 24시간에 4건의 해킹이 발생하므로, 단위 시간은 "시간(hour)"이고 평균 발생 횟수는 4이다.
* 따라서, $λ=4$이며, 확률 밀도 함수는 다음과 같다.

$$f(t) = \frac{4}{24}e^{-\frac{4}{24}t},t≥0$$

- 적분 식을 계산하여 문제를 해결할 수 있다.

$$
\int_{0}^{3} \frac{4}{24} \exp\left(-\frac{4}{24}t\right) dt
$$

$$
= \left[ -\exp\left(-\frac{4}{24}t\right) \right]_0^3
$$

$$
= \exp(0) - \exp\left(-\frac{4}{24} \times 3\right) = 0.3935
$$

## 지수 분포의 특성 - 무기억성

* 지수 분포의 특성으로는 무기억성이 존재한다.
* 특정 시점에서부터 소요되는 시간은 과거로부터 영향을 받지 않는다.
* 예를 들어, 서버가 해킹당하기까지 걸리는 시간은 처음 서버를 운영한 뒤 해킹당하기까지 걸리는 시간과 같다.

→한계점: 현실 세계에서의 다양한 사례를 모델링하기에는 지나치게 단순한 경향이 있다.
