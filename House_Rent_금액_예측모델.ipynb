
 **IQR(Interquartile Range)과 이상치 기준 정리**


>IQR이란?

- IQR = Q3 - Q1
- 중간 50% 데이터 범위를 의미함

---

>이상치(Outlier)의 기준

- 이상치는 너무 작거나 큰 "튀는 값"을 말함
- 아래 기준: **Q1 - 1.5 × IQR**
- 위 기준: **Q3 + 1.5 × IQR**
- 이 범위를 벗어나는 값은 **이상치로 간주**

---

>예시

- Q1 = 20, Q3 = 40 → IQR = 20
- 아래 기준: 20 - (1.5 × 20) = **-10**
- 위 기준: 40 + (1.5 × 20) = **70**

→ 따라서 **-10보다 작거나, 70보다 큰 값은 이상치**

---

>비유

- 박스플롯 = 택배 상자
- 정상 데이터는 상자 안
- 너무 작거나 큰 값은 상자 밖 → **이상치**

# 결측치 처리

```python
rent_df.isna()
```

```python
rent_df.isna().sum()
```

```python
rent_df.isna().mean()
```

```python
rent_df['BHK'] = rent_df['BHK'].fillna(rent_df['BHK'].median())
```

```python
rent_df['Size'] = rent_df['Size'].fillna(rent_df['Size'].mean())
```

```python
rent_df.drop(['Posted On', 'Floor', 'Area Locality'], axis=1, inplace=True)
```

```python
rent_df.head()
```

```python
rent_df = pd.get_dummies(rent_df, columns=['Area Type', 'City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact'],dtype=int)
```

```python
rent_df.head()
```

## 독립변수와 종속변수

###독립변수
 :(Input, Feature, X)
 >결과(종속변수)에 영향을 주는 변수. 우리가 입력하는 데이터

### 종속변수
:(Target, Label, y)
>독립변수의 영향으로 결정되는 결과값. 예측하고자 하는 대상

```python
X = rent_df.drop('Rent', axis=1)
y = rent_df['Rent']
```

```python
# 판다스에서는 대문자는 DataFrame, 소문자는 Series를 표현한다.
X.head(3)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state = 80)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error
```

```python
mean_absolute_error(y_test, pred)
```

```python
mean_squared_error(y_test, pred)
```

```python
root_mean_squared_error(y_test, pred)
```

```python
X_train.loc[1837]
```

```python
X_train.drop(1837, inplace = True)
y_train.drop(1837, inplace = True)
```

```python
lr.fit(X_train, y_train)
```

```python
new_pred = lr.predict(X_test)
```

#Log 활용하기

```python
a = [1,2,3,4,5]
b = [1,10,100,1000,10000]
```

```python
import seaborn as sns
import numpy as np

sns.lineplot(x = a, y = b)
```

```python
b_log = np.log(b)
```

```python
sns.lineplot(x=a, y=b_log)
```

```python
np.exp(b_log)
```

```python
y_train_log=np.log(y_train)
```

```python
y_train_log
```

```python
lr.fit(X_train, y_train_log)
```

```python
newnew_pred = lr.predict(X_test)
```

```python
pred_exp = np.exp(newnew_pred)
```

```python
mean_squared_error(y_test, pred_exp)
```

```python
root_mean_squared_error(y_test,pred_exp)
```
