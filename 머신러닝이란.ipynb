# 머신러닝의 분류

> **supervised leanung**

: 지도학습

타겟 (예측하려는 대상) 이 있어야 하고 그것에 대한 데이터가 있어야한다.

ex) 아파트 가격을 예측함 -> 아파트가격이 곧 타겟


> **unsupervised leaning**

: 비지도학습

타겟값이 지정되지 않는 학습 방법, 어떤것을 예측하려는 목적을 가지지 않음

ex) 클러스터링

- 타겟값이 없기 때문에 컴퓨터가 알아서 데이터를 묶음


> **ReInforcement Leaning**

: 강화학습

보상과 처벌을 기준으로 학습을 함. 올바른 방향으로 나아갈 수 있도록 학습함

수 많은 데이터를 통해 어떤 수를 두었을 때 보상이 되고 처벌이 되는지를 스스로 학습하여

최선의 수를 둘 수 있도록 함.

> **Regression** (회귀)

분류할 수 없는, 일종의 연속된 숫자가 타겟값인 경우

ex) 아파트 가격 예측, 주가예측, 중고차 가격 예측

>**Classification**(분류)

분류할 수 있는 데이터를 가진 타겟값인 경우

ex) 암에 걸릴 환자인지, 이탈할 고객인지 예측하기 -> Y or  N으로 나눌 수 있음ex) 아파트 가격 예측, 주가예측, 중고차 가격 예측



# 머신러닝과 딥러닝의 관계

> 딥러닝은 머신러닝에 포함되고, 머신러닝은 인공지능에 포함됨

**A.I** (인공지능)

인간의 지능을 모방하려는 모든 기술

ex) 머신러닝
<br/>
<br>
**Machine Learning**

데이터를 기반으로 스스로 패턴을 학습하고 예측하는 기술

데이터에서 규칙을 스스로 학습함

ex) 회귀, 의사결정나무, SVM, 랜덤 포레스트 등
<br/>
<br>

**Deep Learning**

인공신경망을 여러 층으로 쌓아 고차원 문제를 해결.

ex) 이미지 처리(CNN), 음성 인식(RNN), 자연어 처리(Transformer, BERT) 등
